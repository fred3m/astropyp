```python
from __future__ import division
import numpy as np

import astropyp

% matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D

def gaussian(amplitude, x, y, x0, y0, sigma_x, sigma_y):
    return amplitude*np.exp(-.5*( (x-x0)**2/sigma_x**2+ (y-y0)**2/sigma_y**2))

np.random.seed(1)

src_count = 20
objs = np.random.random((src_count,2))*100+10
#amplitudes = np.random.random((src_count,))*5000+100
amplitudes = np.random.exponential(500,(src_count,))+100

Y,X = np.indices((100,100))
noise = np.random.poisson(10000,X.size).reshape(X.shape)*1e-3
data = noise
for amp,obj_pos in zip(amplitudes, objs):
    print obj_pos,amp
    g = gaussian(amp,X,Y,obj_pos[1],obj_pos[0],2,2)
    data += g

subset = data[15:36, 30:51]
Xc = np.arange(33,54,1)
Yc = np.arange(13,34,1)
Xc,Yc = np.meshgrid(Xc,Yc)
x_com = np.sum(Xc*subset/np.sum(subset))
y_com = np.sum(Yc*subset/np.sum(subset))
print 'com', x_com, y_com

#plt.imshow(data, cmap='afmhot')
#plt.show()


img = plt.imshow(data,cmap='afmhot') 
## apply norm: 
norm = matplotlib.colors.Normalize(vmin=0, vmax=500) 
img.set_norm(norm)
plt.show()
```

```python
all_new_pos = []
for obj_pos in objs:
    dpos = np.random.rand(2)*3-1.5
    src_pos = (obj_pos[0]+dpos[0], obj_pos[1]+dpos[1])
    patch,X,Y, new_pos = astropyp.utils.misc.get_subpixel_patch(data, src_pos, (21,21), 
        subsampling=5, window_sampling=300, order=5, normalize=False)
    print 'actual', obj_pos
    print 'estimate', src_pos
    print 'subpixel pos', new_pos
    print '***********************'
    all_new_pos.append(new_pos)
```
